# RADIATION HEAT TRANSFER

Calculate the heat transferred via radiation between one cylinder enclosed within another cylinder.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
s=5.67e-8 #Boltzmann constant
d1=50/1000 # Inner diameter in meter
d2=300/1000 # Outer diameter in meter
L=500/1000 # Lenght in meter

A1=np.pi*d1*L # Inner Area
A2=np.pi*d2*L # Outer length

e1=0.04 # Emissivity
e2=0.13

T1=77 # Surface temperature of inner cylinder (T surface)
T2=300 # Surface temperature of outer cylinder (T surrounding)

q=(s*A1*(T1**4-T2**4))/(((1/e1)+((1-e2)/e2)*(d1/d2)))

print(q)
```

```python
s=5.67e-8
d1=50/1000
d2=300/1000
L=500/1000

A1=np.pi*d1*L
A2=np.pi*d2*L

e1=0.04
e2=0.13

T1=np.linspace(4,77,30)
T2=300

q=(s*A1*(T1**4-T2**4))/(((1/e1)+((1-e2)/e2)*(d1/d2)))

plt.plot(T1,q, color= 'red') 
plt.title('Graph of radiative heat rate vs surface temperature')
plt.xlabel('Surface temperature (Kelvin)')
plt.ylabel('Radiative heat rate,q (Watt)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
